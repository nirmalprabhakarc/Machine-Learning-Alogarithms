```python
# Importing necessary libraries
import pandas as pd
from itertools import combinations
print("URK21CS2060")
```

```python
#1. Read the data
data = pd.read_csv("/content/B5_Shop2.csv")
data
print("URK21CS2060")
```

```python
#2. Display the candidate set and frequency set for every iteration
print("URK21CS2060")
from itertools import combinations
def generate_initial_set(data):
    transactions = list(data['Item'].apply(lambda x: set(x.split(','))))
    return transactions

def create_candidate_set(transactions, size):
    itemsets = set()
    for transaction in transactions:
        for itemset in combinations(transaction, size):
            itemsets.add(itemset)
    return itemsets

def calculate_support(transactions, candidate_set):
    support_count = {itemset: 0 for itemset in candidate_set}
    for transaction in transactions:
        for itemset in candidate_set:
            if set(itemset).issubset(transaction):
                support_count[itemset] += 1
    return support_count

def apriori(transactions, min_support):
    result = []
    size = 1
    all_frequent_itemsets = {}
    while True:
        candidate_set = create_candidate_set(transactions, size)
        if not candidate_set:
            break
        support_count = calculate_support(transactions, candidate_set)
        frequent_itemsets = {itemset: count/len(transactions) for itemset, count in support_count.items() if count/len(transactions) >= min_support}
        if not frequent_itemsets:
            break
        result.append(frequent_itemsets)
        all_frequent_itemsets.update(frequent_itemsets)
        size += 1

    print("Candidate Sets and Frequency Sets:")
    print("Support   Itemsets")

    for i, (itemset, support) in enumerate(all_frequent_itemsets.items(), 1):
        if i <= 10 or i > len(all_frequent_itemsets) - 10:
            print(f"{i}  {support:.2f}  {itemset}")
        elif i == 11:
            print("... ...")

    return result
transactions = generate_initial_set(data)
min_support = 0.08  # Minimum support threshold
frequency_sets = apriori(transactions, min_support)
```

```python
#3. Display the association rules
print("URK21CS2060")
def generate_rules(frequency_sets, transactions, min_confidence):
    rules = []
    for frequent_itemsets in frequency_sets:
        for itemset in frequent_itemsets.keys():
            for size in range(1, len(itemset)):
                for antecedent in combinations(itemset, size):
                    consequent = set(itemset) - set(antecedent)
                    antecedent_support = calculate_support(transactions, [antecedent])[antecedent]
                    rule_support = frequent_itemsets[itemset] * len(transactions)
                    confidence = rule_support / antecedent_support
                    if confidence >= min_confidence:
                        rules.append(((antecedent, tuple(consequent)), confidence))
    return rules

num_rules_to_display = 30
min_confidence = 0.7
association_rules = generate_rules(frequency_sets, transactions, min_confidence)
for i, rule in enumerate(association_rules,1):
  if i <= 10 or i > len(association_rules) - 10:
    print(f"Rule: {rule[0][0]} -> {rule[0][1]}, Confidence: {rule[1]}")
  elif i == 11:
    print("... ...")
```

```python
#4. Find all the rules of these subsets that have higher confidence value
print("URK21CS2060")
transactions = generate_initial_set(data)
min_support = 0.08  # Minimum support threshold
min_confidence = 0.5  # Minimum confidence threshold
association_rules = generate_rules(frequency_sets, transactions, min_confidence)
for rule in association_rules[:5]:
    print(f"Rule: {rule[0][0]} -> {rule[0][1]}, Confidence: {rule[1]}")
```

# Linear Regression


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Cervical Cancer - Cervical Cancer.csv')
df
```

```python
x = df['behavior_eating']
y = df['behavior_personalHygine']
```

```python
xbar = sum(x)/len(x)
ybar = sum(y)/len(y)
```

```python
num = sum((x - xbar)*(y - ybar))
deno = sum((x - xbar)**2)
```

```python
b1 = num/deno
b1
```

```python
b0 = ybar - b1 * xbar
b0
```

```python
y_pred = b0 + b1*x
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred, color = 'red')
```

# Logistic

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```
